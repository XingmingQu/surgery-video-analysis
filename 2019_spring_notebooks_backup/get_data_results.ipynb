```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
from matplotlib.patches import Rectangle
import warnings
warnings.filterwarnings('ignore')
```

```python
def convert_to_df(data,colnum):
    loss=0
    result=[]    
    for i in range(len(data)):
        loss+=np.sum(data[i][colnum])
        result.append(data[i][colnum])
 
    result=pd.DataFrame(result)
    result.columns = ['DP','BD','E','FS','A','RC']
    return loss,result

def getlabel(number):
    L=Video_label[Video_label['label']==number]
    L=np.array(L)[:,:-1]
    L= np.squeeze(L)
    return L

range_label=pd.read_csv('range_label.csv')
Video_label=pd.read_csv('newall_label.csv')
range_label.head()
counts=range_label['VideoNum'].value_counts()

def get_review_times(videoNum):
    counts=range_label['VideoNum'].value_counts()
    times=int(counts[counts.index==videoNum])
    return times
```

```python
r_group=range_label.groupby(['VideoNum'])
r_max=r_group.max()
r_min=r_group.min()
r_max=r_max.reset_index()
r_min=r_min.reset_index()
def get_video_range(videoNum):
    vmax=r_max[r_max['VideoNum']==videoNum]
    vmin=r_min[r_min['VideoNum']==videoNum]
    return vmax,vmin
```

```python
def number_of_score_in_range(re,gt,ran):
    count=0
    for i,j in zip(re,gt):
        if abs(float(i)-float(j))<=ran:
            count+=1
    return count
```

```python
best=np.load('data\Diff19_0.01_0.1_0.333333_0.333333_0.333333_fsize6.npy')
best_residual=np.load('data\Residual19_0.01_0.1_0.333333_0.333333_0.333333_fsize6.npy')

best_loss,best_res=convert_to_df(best,3)
_,residual=convert_to_df(best_residual,3)
_,residual_noround=convert_to_df(best_residual,2)
print(best_loss)
```

```python
Y=[11, 113, 130, 143, 152, 16, 164, 171, 176, 192, 194, 207, 236, 237, 240, 253, 257, 258, 49, 59, 74, 78, 91]
# Y=[11, 113,130]

raw_final_valid_score_1=0
raw_final_valid_score_2=0
round_final_valid_score_1=0
round_final_valid_score_2=0
for i,index in zip(Y,range(len(Y))):
    print("Video Number:",i)
    print("This video has %d reviewer"%(get_review_times(i)))
    ax=plt.figure(figsize=(18,3))
    
    gt=getlabel(i)
    re=gt+residual.iloc[index]
    re_none_round=gt+residual_noround.iloc[index]

    
    y=[x for x in range(1,7)]
    ax = plt.subplot(141)
#     Plot_data.plot.box(ax=ax)
#     ax.plot(y, re, label='predict-round',color='red')
    ax.plot(y, re_none_round, label='predict',color='blue')
    
    ax.plot(y, gt, label='ground truth',color='green')

    plt.ylabel('Gear Score')
    plt.xlabel('Score type')
    plt.ylim(0,5.5)
    ax.legend()
    
    ax = plt.subplot(142)
    times=get_review_times(i)
#     ax.plot(y, np.round(fi_re), label='predict-round')
    ax.plot(y, re, label='predict-round',color='red')
    ax.plot(y, re_none_round, label='predict',color='blue')
    plt.ylabel('Gear Score')
    plt.xlabel('Score type')
    plt.ylim(0,5.5)

    if times>1:        
        vmax,vmin=get_video_range(i)

        it=1
        for types in ['DP','BD','E','FS','A','RC']:
            Range=vmax[types]-vmin[types]
            rect = Rectangle((it, vmin[types]), 0.2, Range, color='green')
            ax.add_patch(rect)   
            it=it+1
    else:
    #             new_re=np.round(video_result.iloc[3])
    #             ax.plot(y, new_re, label='predict-round')
        ax.plot(y, gt, label='ground truth')        
    labels=['','DP','BD','E','FS','A','RC']
    ax.set_xticklabels(labels)
    ax.legend()            
    
    
    gt = [ '%.5f' % elem for elem in gt ]
    print("The groud truth of this video",gt)
    re_none_round = [ '%.5f' % elem for elem in re_none_round ]
    print("Our raw prediction           ",list(re_none_round))
    
    re = [ '%.5f' % elem for elem in re ]
    print("Our round prediction         ",list(re))
    
#     print(number_of_score_in_range(re_none_round,gt,1))
    print("For raw prediction:")
    print("    %d/6 prediciton are in +-%d range" %(number_of_score_in_range(re_none_round,gt,1),1))
    print("    %d/6 prediciton are in +-%d range" %(number_of_score_in_range(re_none_round,gt,2),2))
    print("    %d/6 prediciton are in +-%d range" %(number_of_score_in_range(re_none_round,gt,3),3))

    print("For rounded prediction:")
    print("    %d/6 prediciton are in +-%d range" %(number_of_score_in_range(re,gt,1),1))
    print("    %d/6 prediciton are in +-%d range" %(number_of_score_in_range(re,gt,2),2))
    print("    %d/6 prediciton are in +-%d range" %(number_of_score_in_range(re,gt,3),3))
    
    raw_final_valid_score_1+=number_of_score_in_range(re_none_round,gt,1)
    raw_final_valid_score_2+=number_of_score_in_range(re_none_round,gt,2)

    round_final_valid_score_1+=number_of_score_in_range(re,gt,1)
    round_final_valid_score_2+=number_of_score_in_range(re,gt,2)
    
    plt.show()
```
