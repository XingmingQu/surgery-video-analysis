```python
def getVideoLength(name):
    cap = cv2.VideoCapture(name)
    fps = cap.get(cv2.CAP_PROP_FPS)      # OpenCV2 version 2 used "CV_CAP_PROP_FPS"
    frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    duration = frame_count/fps

#     print('fps = ' + str(fps))
#     print('number of frames = ' + str(frame_count))
#     print('duration (S) = ' + str(duration))
#     minutes = int(duration/60)
#     seconds = duration%60
#     print('duration (M:S) = ' + str(minutes) + ':' + str(seconds))
#     cap.release()
    return int(duration)
```

```python
csvfile = '..\\interRaterReliability\\interRaterReliabilityLabels.csv'
name = "D:\\surgeryVideoData\\Segmented_videoClips\\"

df = pd.read_csv(csvfile)
df.head(10)
```

```python
df.length.unique()
```

```python

def get_interRaterReliabilityForTwoReviewer(r1,r2,groupedDf):
    reviewVideos1 = groupedDf[groupedDf['Reviewer'] == r1].VideoNum.unique()
    reviewVideos2 = groupedDf[groupedDf['Reviewer'] == r2].VideoNum.unique()
    

    intersect = np.intersect1d(reviewVideos1,reviewVideos2)
#     print(reviewVideos1,reviewVideos2,intersect)
    if len(intersect) == 0:
        return "0/0"
    
    totaloverlap = len(intersect)
#     #for each video check if matches or not
    matches = 0
    for video in intersect:
#         print(video)
        score1 = np.mean(groupedDf[(groupedDf['Reviewer'] == r1) & (groupedDf['VideoNum'] == video)].sumScore)
        score2 = np.mean(groupedDf[(groupedDf['Reviewer'] == r2) & (groupedDf['VideoNum'] == video)].sumScore)
#         print(score1,score2)
        if abs(score1-score2)<=2:
            matches = matches+1
        
        re= "{}/{} = {:.3f}".format(matches,totaloverlap,matches/totaloverlap)
#         print(re)
    return re

def get_interRaterReliability_matrix_for_videoGroup(video_length):
    print("Video Length:",video_length)
    groupedDf = df[df['length'] == video_length]
    reviewers = groupedDf.Reviewer.unique()
#     print(reviewers)

    matrix =  [['NA' for _ in range(len(reviewers))] for _ in range(len(reviewers))]
    for i in range(len(reviewers)-1):
        for j in range(i+1,len(reviewers)):
#             print(i,j)
            matrix[i][j]=get_interRaterReliabilityForTwoReviewer(reviewers[i],reviewers[j],groupedDf)

    result_df = pd.DataFrame(matrix,columns=reviewers)
    return result_df
```

```python
for i in df.length.unique():
    print(get_interRaterReliability_matrix_for_videoGroup(i))
    print('\n')
```
